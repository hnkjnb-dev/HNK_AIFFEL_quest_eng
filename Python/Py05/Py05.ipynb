```python
import random

# Q1. Account 클래스 생성 : 은행이름, 예금주, 계좌번호, 잔액 설정
class Account:
  account_count = 0

  def __init__(self, holder, number, balance):
    self.holder = holder
    self.number = number
    self.balance = balance
    self.deposit_count = 0

    self.deposit_log = []
    self.withdraw_log = []

    # Q2. 클래스 변수를 사용해서 Account 클래스로부터 생성된 계좌 객체의 개수를 저장
    Account.account_count += 1
    print("총 계좌의 수 : ", Account.account_count)

    # 랜덤 계좌번호 생성(3-2-6)
    part_1 = str(random.randint(0, 999))
    part_2 = str(random.randint(0, 99))
    part_3 = str(random.randint(0, 999999))
    self.account_number = part_1 + '-' + part_2 + '-' + part_3

  # Q3. 생성된 계좌의 개수를 출력하는 get_account_num 메서드 추가
  def get_account_num(self):
    return Account.account_count

  # Q4. 입금을 위한 deposit 메서드 추가 (입금은 최소 1원 이상만 가능)
  def deposit(self, amount):
    if amount >= 1:
      self.balance += amount
      self.deposit_count += 1
      self.deposit_log.append(amount)
      print(f"{amount}원이 입금되었습니다.")
    else:
      print("1원 이상만 입금 가능합니다.")
      return

    # Q7. 입금 횟수 5회마다 잔고 기준 1%의 이자 지급
    if self.deposit_count % 5 == 0:
        self.balance = int(self.balance * 1.01)
        print("입금 5회 누적으로 이자 1%가 지급되었습니다.")

  # Q5. 출금을 위한 withdraw 메서드 추가
  def withdraw(self, amount):
    if self.balance >= amount:
      self.balance -= amount
      self.withdraw_log.append(amount)
      print(f"{amount}원이 출금되었습니다.")
    else:
      print("잔액이 부족합니다.")
  
  # Q6. 정보 출력 메서드 : 
  def display_info(self):
    print("은행이름 : SC은행", ", "
          "예금주 : ", self.holder, ", ",
          "계좌번호 : ", self.account_number,", ",
          "잔고 : ", self.balance, "원")
  
  # Q10. 입금과 출금 내역이 기록되도록 코드를 업데이트
  def deposit_history(self):
    print(self.deposit_log)
  def withdraw_history(self):
    print(self.withdraw_log)
```

```python
### Quest 1. Chat GPT 코드

import random
from datetime import datetime

class Account:
    # Q2: 클래스 변수로 계좌 개수 저장
    account_count = 0

    def __init__(self, owner: str, balance: int = 0):
        # Q1: 은행이름 고정, 계좌번호 랜덤 생성, 예금주/초기잔액 입력
        self.bank = "SC은행"
        self.owner = owner
        self.account_number = self._generate_account_number()
        self.balance = int(balance)

        # Q7: 입금 횟수 카운트
        self._deposit_count = 0

        # Q10: 입출금 내역 기록용
        # (원하면 문자열만 저장해도 되지만, 날짜/금액/잔액을 같이 저장하면 더 유용)
        self._deposit_logs = []   # list[dict]
        self._withdraw_logs = []  # list[dict]

        # Q2: 생성될 때마다 개수 증가
        Account.account_count += 1

    @staticmethod
    def _generate_account_number() -> str:
        # 3자리-2자리-6자리
        part1 = f"{random.randint(0, 999):03d}"
        part2 = f"{random.randint(0, 99):02d}"
        part3 = f"{random.randint(0, 999999):06d}"
        return f"{part1}-{part2}-{part3}"

    # Q3: 계좌 개수 출력 메서드
    @classmethod
    def get_account_num(cls) -> int:
        return cls.account_count

    # Q4: 입금 (최소 1원 이상)
    def deposit(self, amount: int) -> bool:
        amount = int(amount)
        if amount < 1:
            return False

        self.balance += amount
        self._deposit_count += 1

        # Q10: 입금 내역 기록
        self._deposit_logs.append({
            "time": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "amount": amount,
            "balance": self.balance
        })

        # Q7: 입금 5회마다 1% 이자 지급
        if self._deposit_count % 5 == 0:
            interest = int(self.balance * 0.01)  # 원 단위 정수 처리
            if interest > 0:
                self.balance += interest
                # 이자도 입금 내역으로 기록(표시용)
                self._deposit_logs.append({
                    "time": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                    "amount": interest,
                    "balance": self.balance,
                    "type": "interest"
                })

        return True

    # Q5: 출금 (잔고 초과 불가)
    def withdraw(self, amount: int) -> bool:
        amount = int(amount)
        if amount < 1:
            return False
        if amount > self.balance:
            return False

        self.balance -= amount

        # Q10: 출금 내역 기록
        self._withdraw_logs.append({
            "time": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "amount": amount,
            "balance": self.balance
        })
        return True

    # Q6: 정보 출력 (잔고 천단위 콤마)
    def display_info(self) -> None:
        print(
            f"(은행이름: {self.bank}, 예금주: {self.owner}, "
            f"계좌번호: {self.account_number}, 잔고: {self.balance:,}원)"
        )

    # Q10: 입금/출금 내역 출력 메서드
    def deposit_history(self) -> None:
        print(f"[{self.owner}] 입금 내역")
        if not self._deposit_logs:
            print("  (없음)")
            return
        for log in self._deposit_logs:
            log_type = log.get("type", "deposit")
            tag = "이자" if log_type == "interest" else "입금"
            print(f"  {log['time']} | {tag}: {log['amount']:,}원 | 잔고: {log['balance']:,}원")

    def withdraw_history(self) -> None:
        print(f"[{self.owner}] 출금 내역")
        if not self._withdraw_logs:
            print("  (없음)")
            return
        for log in self._withdraw_logs:
            print(f"  {log['time']} | 출금: {log['amount']:,}원 | 잔고: {log['balance']:,}원")


# -----------------------------
# Q8: 3개 이상 객체 생성 후 리스트에 저장
accounts = [
    Account("파이썬", 1_500_000),
    Account("자바", 900_000),
    Account("C++", 2_300_000),
]

# Q3: 생성된 계좌 개수 출력
print("생성된 계좌 수:", Account.get_account_num())

# Q9: 잔고가 100만원 이상인 고객 정보만 출력
print("\n[잔고 100만원 이상 고객]")
for acc in accounts:
    if acc.balance >= 1_000_000:
        acc.display_info()

# (테스트) Q4~Q7, Q10 동작 확인
print("\n[입금 5회 테스트: 5회마다 이자 1% 지급]")
a = accounts[0]
for _ in range(5):
    a.deposit(10_000)  # 5회째에 이자 발생

a.withdraw(50_000)

print("\n[정보 출력]")
a.display_info()

print("\n[내역 출력]")
a.deposit_history()
a.withdraw_history()
```

```python
### Quest 2. Chat GPT 코드
# Quest 2.는 아직 직접 코드 작성 이전으로, 주말 동안 학습할 예정입니다.

import random

class Character:
    def __init__(self, name: str, level: int, hp: int, atk: int, defense: int):
        self.name = name
        self.level = level
        self.hp = hp
        self.atk = atk
        self.defense = defense

    # Q1: 현재 체력이 0 이상인지
    def is_alive(self) -> bool:
        return self.hp > 0

    # Q1: (받은 데미지 - 방어력) 만큼 체력 감소 (방어력이 더 크면 감소 X)
    def take_damage(self, damage: int) -> int:
        real_damage = damage - self.defense
        if real_damage <= 0:
            real_damage = 0
        self.hp -= real_damage
        return real_damage

    # Q1: 1~공격력 사이 랜덤 데미지로 공격
    def attack_target(self, target: "Character") -> int:
        damage = random.randint(1, self.atk)
        target.take_damage(damage)
        return damage


class Player(Character):
    def __init__(self, name: str):
        # Q2: 레벨 1, 체력 100, 공격력 25, 방어력 5
        super().__init__(name=name, level=1, hp=100, atk=25, defense=5)
        self.experience = 0

    # Q2: 경험치 획득
    def gain_experience(self, amount: int) -> None:
        self.experience += int(amount)

    # Q2: 경험치가 50 이상이면 레벨업 (요구사항 그대로: 한 번에 1회 상승)
    def level_up(self) -> None:
        if self.experience >= 50:
            self.level += 1
            self.atk += 10
            self.defense += 5
            # 보통은 경험치를 깎지만(예: -50), 문제에 명시가 없어서 유지함
            # 원하면 self.experience -= 50 로 바꿔도 됨


class Monster(Character):
    def __init__(self, level: int):
        level = int(level)

        # Q2: 레벨 비례 스탯
        hp = random.randint(10, 30) * level
        atk = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level

        super().__init__(name=f"몬스터(Lv.{level})", level=level, hp=hp, atk=atk, defense=defense)


# Q3: battle 함수
def battle(player: Player, monster: Monster) -> None:
    print(f"\n=== 전투 시작: {player.name}(Lv.{player.level}) vs {monster.name} ===")

    # 누가 먼저 치는지 랜덤으로 해도 되지만, 문제에 없으니 플레이어 선공으로 진행
    while player.is_alive() and monster.is_alive():
        # 플레이어 공격
        p_damage = player.attack_target(monster)
        print(f"{player.name}의 공격! (공격력 기반 데미지 {p_damage}) -> "
              f"{monster.name} HP: {max(monster.hp, 0)}")

        if not monster.is_alive():
            break

        # 몬스터 공격
        m_damage = monster.attack_target(player)
        print(f"{monster.name}의 공격! (공격력 기반 데미지 {m_damage}) -> "
              f"{player.name} HP: {max(player.hp, 0)}")

    if player.is_alive():
        gained = monster.level * 20
        player.gain_experience(gained)
        player.level_up()
        print("전투 승리!")
        print(f"획득 경험치: {gained}, 현재 경험치: {player.experience}, 현재 레벨: {player.level}")
    else:
        print("전투 패배..")


# ----------------- 실행 예시(원하면 제거) -----------------
if __name__ == "__main__":
    name = input("플레이어 이름을 입력하세요: ").strip() or "용사"
    player = Player(name)

    # 자동사냥: 몬스터를 차례대로 (예시로 3마리)
    for lvl in [1, 2, 3]:
        if not player.is_alive():
            break
        monster = Monster(lvl)
        battle(player, monster)

    print("\n게임 종료")
    print(f"{player.name} - Lv.{player.level}, HP:{max(player.hp,0)}, ATK:{player.atk}, DEF:{player.defense}, EXP:{player.experience}")
```
